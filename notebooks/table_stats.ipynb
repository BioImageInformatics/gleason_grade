```python
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('seaborn-white')
import seaborn as sns

from scipy import stats
```

```python
# x = np.array([21.8, 84.9, 58.3, 0.76])
x = np.array([0.752, 0.769, 0.665, 0.719])
y = np.array([0.674, 0.81, 0.81, 0.813])
print(stats.spearmanr(x, y))
sns.regplot(x, y)
```

```python
densenet = np.array([0.75, 0.767, 0.719])
fcn = np.array([0.712, 0.738, 0.748])
unet = np.array([0.771, 0.803, 0.673])

densenet_s = np.array([0.749, 0.805, 0.726])
fcn_s = np.array([0.699, 0.74, 0.74])
unet_s = np.array([0.767, 0.785, 0.801])

print(stats.f_oneway(np.concatenate([densenet, fcn, unet]), np.concatenate([densenet_s, fcn_s, unet_s])))
print(stats.f_oneway(densenet, densenet_s))
print(stats.f_oneway(fcn, fcn_s))
print(stats.f_oneway(unet, unet_s))
```

```python
x = np.array([0.75, 0.767, 0.719, 0.712, 0.738, 0.748, 0.771, 0.803, 0.673, 0.749, 0.805, 0.726, 0.699, 0.74, 0.74, 0.767, 0.785, 0.801])
y = np.array([0.763, 0.748, 0.636, 0.775, 0.745, 0.676, 0.806, 0.797, 0.429, 0.794, 0.845, 0.649, 0.725, 0.715, 0.653, 0.733, 0.692, 0.626])
print(stats.ttest_rel(x, y))
print(stats.spearmanr(x, y))
fig, ax = plt.subplots(1,1, figsize=(3,3))
sns.regplot(x, y, fit_reg=False, ax=ax)
plt.plot((0,1), (0,1))
plt.xlim([0.4, 0.9])
plt.ylim([0.4, 0.9])
plt.title('Constant FOV\nVarying Magnification', fontsize=20)
plt.xlabel('Tile F1', fontsize=18)
plt.ylabel('WSI F1', fontsize=18)
fig.savefig('/home/nathan/Dropbox/_papers/gleason_grade/mag_expt_wsi_vs_tile.pdf', bbox_inches='tight')
```
