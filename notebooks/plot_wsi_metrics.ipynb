```python
## evaluate majority correct

y_true = table['Label'].values
y_hat = []
y_pct = []

for idx in table.index:
    y_ = table.loc[idx, 'Label']
    counts = table.iloc[idx, 3:7].values
    if y_ == 2:
        counts[2] = counts[1] + counts[2]
        counts[1] = 0
    pred = np.argmax(counts)
    y_hat.append(pred)
    
    pcts = counts / np.sum(counts).astype(np.float32)
#     y_pct.append(pcts[y_])
    y_pct.append(np.expand_dims(pcts, 0))
    
y_hat = np.asarray(y_hat)
y_pct = np.concatenate(y_pct, axis=0)
print(classification_report(y_true=y_true, y_pred=y_hat))

```
