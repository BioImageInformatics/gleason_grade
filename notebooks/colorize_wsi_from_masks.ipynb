```python
masks = glob.glob('../data/validation_svs/*png')
maskbases = [os.path.basename(maskpath).replace('.png','') for maskpath in masks]
masks = {maskbase: maskpath for maskbase, maskpath in zip(maskbases, masks)}
```

```python
img5x = glob.glob('../densenet_small/ext_10pct/inference/*_rgb.jpg')
img5xbases = [os.path.basename(img5xpath).replace('_rgb.jpg', '') for img5xpath in img5x]
img5x = {img5xbase: img5xpath for img5xbase, img5xpath in zip(img5xbases, img5x)}
```

```python
coeff = 1.2
outdir = '../data/validation_svs'
def color_by_mask(maskpath, imgpath, basename):
    mask = cv2.imread(maskpath, -1)
    img = cv2.imread(imgpath, -1)[:,:,::-1]
    
    r,g,b = np.split(img, 3, axis=-1)
#     r = r.astype(np.float32)
#     g = g.astype(np.float32)
#     b = b.astype(np.float32)
    
    mask = cv2.resize(mask, dsize=img.shape[:2][::-1], interpolation=cv2.INTER_NEAREST)
    outpath = os.path.join(outdir, '{}_mask_colored.jpg'.format(basename))
    
    classes = np.unique(mask)
    classes = classes[classes != 255]
    
    for c in classes:    
        cmask = mask == c
#         r[cmask] *= coeff*colors[c, 0]/255.
#         g[cmask] *= coeff*colors[c, 1]/255.
#         b[cmask] *= coeff*colors[c, 2]/255.
        if c == 2: c=1
        r[cmask] = colors[c, 0]
        g[cmask] = colors[c, 1]
        b[cmask] = colors[c, 2]
    
#     r = r.astype(np.uint8)
#     g = g.astype(np.uint8)
#     b = b.astype(np.uint8)
    imgout = np.dstack([b,g,r])
    cv2.imwrite(outpath, imgout)
    print(mask.shape, img.shape, outpath, classes)
    
```

```python
for key in masks.keys():
    try:
        color_by_mask(masks[key], img5x[key], key)
    except Exception as e:
        print(e)
        print(key)
```
